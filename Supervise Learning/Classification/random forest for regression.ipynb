```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
from sklearn.tree import plot_tree
```

```python
#load the boston housing dataset
boston = load_boston()
```

```python
boston
```

```python
boston.data #independent variables
```

```python
boston.target #dependent variables
```

```python
df = pd.DataFrame(data = boston.data,columns = boston.feature_names)
df['PRICE'] =boston.target
```

```python
df.head()
```

```python
df.info()
```

```python
df.tail()
```

```python
df.isnull().sum()
```

```python
X = df.iloc[:,:-1]
X
```

```python
y = df.iloc[:,-1]
y
```

```python
#train-test-split
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 42)
```

```python
X_train
```

```python
y_train
```

```python
#Initializint the random forest regressor
from sklearn.ensemble import RandomForestRegressor 
```

```python
#Initializes a Random Forest Regressor with 100 trees and trains it on the training set.
rfg = RandomForestRegressor(n_estimators =100, random_state = 42) 
```

```python
#fit the model

rfg.fit(X_train, y_train)
```

```python
#prediction

y_pred = rfg.predict(X_test)
```

```python
y_pred
```

```python
#evaluation

from sklearn.metrics import mean_squared_error, r2_score
```

```python
mse = mean_squared_error(y_test, y_pred)
mse
```

```python
r2 = r2_score(y_test, y_pred)
r2
```

```python

# Feature importance
feature_importances = rfg.feature_importances_
importances_df = pd.DataFrame({
    'Feature': X.columns,
    'Importance': feature_importances
}).sort_values(by='Importance', ascending=False)

print("Feature Importances:")
print(importances_df)
```

```python
estimator = rfg.estimators_[0]
```

```python
plt.figure(figsize = (20,10))

plot_tree(estimator, filled=True, feature_names=X.columns)
plt.show()
```

```python
#scatter plot

plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.6)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--')
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Actual vs Predicted Prices')
plt.show()

```

```python

# Sort the DataFrame by importance
importances_df = importances_df.sort_values(by='Importance', ascending=True)

# Plot
plt.figure(figsize=(10, 6))
plt.barh(importances_df['Feature'], importances_df['Importance'], color='skyblue')
plt.xlabel('Importance Score')
plt.ylabel('Feature')
plt.title('Feature Importances in Random Forest Model')
plt.show()
```
