```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#boston house pricing dataset

from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
boston
```

```python
print(boston.data) #independent data 
print(boston.target) #dependent data 
print(boston.feature_names) #columns
```

```python
X = pd.DataFrame(boston.data, columns = boston.feature_names)
X
```

```python
y = boston.target
y
```

```python
#train_test_split

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 42)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
regressor = DecisionTreeRegressor()
```

```python
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
y_pred
```

```python
#r2 score = a statistical measure in a regression model that shows how well the data fits the model

from sklearn.metrics import r2_score
score = r2_score(y_pred,y_test)
score
```

```python
# Hyperparameter Tuning
parameter = {
    'criterion': ['squared_error', 'frieman_mse', 'absolute_error', 'poisson'],
    'splitter': ['best', 'random'],
    'max_depth': [1, 2, 3, 4, 5,6,7,8, 10, 11,12,13,15],
    'max_features': ['auto', 'sqrt', 'log2']
}
```

```python
regressor = DecisionTreeRegressor()
```

```python
#GridSearchCV acts as a valuable tool for identifying the optimal parameters for a machine learning model. 
from sklearn.model_selection import GridSearchCV 
```

```python
regressorCV = GridSearchCV(regressor, param_grid = parameter, cv = 5, scoring ='neg_mean_squared_error' )
```

```python
regressorCV.fit(X_train, y_train)
```

```python
regressorCV.best_params_
```

```python
y_pred = regressorCV.predict(X_test)
y_pred
```

```python
r2_score(y_pred,y_test)
```
