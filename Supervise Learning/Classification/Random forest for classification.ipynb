```python
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
```

```python
iris
```

```python
iris.data
```

```python
iris.target
```

```python
df = pd.DataFrame(data = iris.data, columns = iris.feature_names)
df['species']= iris.target
```

```python
df
```

```python
X = df.iloc[:,:-1]
X
```

```python
y = df.iloc[:,-1]
y
```

```python
#train-test-split

from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators = 100, random_state = 42)
```

```python
#fit the model on training data

rfc.fit(X_train, y_train)
```

```python
#make predictions on the test data

y_pred = rfc.predict(X_test)
```

```python
y_pred
```

```python
#Evaluate the model

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
classification_rep = classification_report(y_test, y_pred)
print(classification_rep)
```

```python
#feature importance


feature_importance = rfc.feature_importances_

importance_df = pd.DataFrame({
    'Feature': X.columns,
    'Importance': feature_importance
}).sort_values(by = 'Importance')

print("feature importance :")
importance_df

```

```python
# plotting feature importance


plt.figure(figsize = (10, 6))
plt.bar(importance_df['Feature'], importance_df['Importance'], color ='indigo')
plt.xlabel('Features')
plt.ylabel('Importance')
plt.title('Feature Importance')
plt.xticks(rotation = 40)
```

```python
from sklearn.tree import plot_tree

# Visualizing the first tree in the forest
plt.figure(figsize=(20, 10))
plot_tree(rfc.estimators_[0], filled=True, feature_names=X.columns, class_names=iris.target_names)
plt.show()

```

```python
plt.figure(figsize=(20, 10))
plot_tree(rfc.estimators_[1], filled=True, feature_names=X.columns, class_names=iris.target_names)
plt.show()
```
