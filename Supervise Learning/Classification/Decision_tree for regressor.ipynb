```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline # Plots are rendered directly within the notebook's output cells, so you don't need to open a separate window to view them.
```

```python
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
boston
```

```python
boston.data
```

```python
boston.target
```

```python
df = pd.DataFrame(data = boston.data,columns = boston.feature_names)
df['PRICE'] =boston.target
```

```python
df.head()
```

```python
X = df.iloc[:,:-1]
X
```

```python
y = df.iloc[:,-1]
y
```

```python
#train_test_split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
regressor=DecisionTreeRegressor()
```

```python
regressor.fit(X_train, y_train)
```

```python

y_pred=regressor.predict(X_test)

y_pred
```

```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(y_pred,y_test)
score
```

```python

## Hyperparameter Tunning
parameter={
 'criterion':['squared_error','friedman_mse','absolute_error','poisson'],
  'splitter':['best','random'],
  'max_depth':[1,2,3,4,5,6,7,8,10,11,12],
  'max_features':['auto', 'sqrt', 'log2']
    
}
regressor=DecisionTreeRegressor()
```

```python
#https://scikit-learn.org/stable/modules/model_evaluation.html
from sklearn.model_selection import GridSearchCV
regressorcv=GridSearchCV(regressor,param_grid=parameter,cv=5,scoring='neg_mean_squared_error')
```

```python

regressorcv.fit(X_train,y_train)
```

```python
regressorcv.best_params_
```

```python
y_pred = regressorcv.predict(X_test)
y_pred
```

```python
r2_score(y_pred,y_test)
```

```python
# Get the best score
best_score = -regressorcv.best_score_  # Negate because GridSearchCV uses negative MSE
print("Best mean squared error found: ", best_score)
```

```python
# Use the best model to make predictions
best_regressor = regressorcv.best_estimator_
y_pred_best = best_regressor.predict(X_test)
y_pred
```

```python
# Evaluate the best model
score_best = r2_score(y_test, y_pred_best)
print("R² score of the best model: ", score_best)
```
