```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris
```

```python
iris.data #independent variable
```

```python
iris.target #dependent variable
```

```python
iris.feature_names # column names
```

```python
X = pd.DataFrame(iris.data,columns = iris.feature_names)
X
```

```python
y = iris.target
y
```

```python
#train_test_split

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 42)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
#by default it create postpruning technique

classifier = DecisionTreeClassifier()
```

```python
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
y_pred
```

```python
# Hyperparameter Tuning
parameter = {
    'criterion': ['squared_error', 'frieman_mse', 'absolute_error', 'poisson'],
    'splitter': ['best', 'random'],
    'max_depth': [1, 2, 3, 4, 5,6,7,8, 10, 11,12,13,15],
    'max_features': ['auto', 'sqrt', 'log2']
}
```

```python
classifier = DecisionTreeClassifier()
```

```python
#GridSearchCV acts as a valuable tool for identifying the optimal parameters for a machine learning model. 
from sklearn.model_selection import GridSearchCV 
```

```python
classifiercv = GridSearchCV(classifier, param_grid = parameter, cv = 5, scoring ='neg_mean_squared_error' )
```

```python
classifiercv.fit(X_train,y_train)
```

```python
classifiercv.best_params_
```
