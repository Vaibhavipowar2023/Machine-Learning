# Gaussian Naive Bayes

GaussianNB is specifically used when the features have continuous values

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris
```

```python
 iris.data
```

```python
iris.target
```

```python
df = pd.DataFrame(data = iris.data,columns =iris.feature_names)

df['Species'] =iris.target
```

```python
df
```

```python
X = df.iloc[:,: -1] #dependent variable
X
```

```python
y = df.iloc[:,-1] #independent variable
y
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
#GaussianNB is specifically used when the features have continuous values

from sklearn.naive_bayes import GaussianNB
```

```python
model = GaussianNB()
```

model

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
#chech the accuracy


from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)
accuracy
#gives 100% accuracy
```

```python
#chech confusion metrics

from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, y_pred)
```

```python

from matplotlib.colors import ListedColormap


x_set, y_set = X_train, y_train

# Create a meshgrid for plotting
X1, X2 = np.meshgrid(np.arange(start = x_set.iloc[:, 0].min() - 1, stop = x_set.iloc[:, 0].max() + 1, step = 0.01),  
                     np.arange(start = x_set.iloc[:, 1].min() - 1, stop = x_set.iloc[:, 1].max() + 1, step = 0.01))  

# Plot the decision boundary using the meshgrid
plt.contourf(X1, X2, model.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),  
             alpha = 0.75, cmap = ListedColormap(('red', 'green', 'blue')))  

# Plot the data points
plt.xlim(X1.min(), X1.max())  
plt.ylim(X2.min(), X2.max())  

for i, j in enumerate(np.unique(y_set)):  
    plt.scatter(x_set.iloc[y_set == j, 0], x_set.iloc[y_set == j, 1],  
                c = ListedColormap(('red', 'green', 'blue'))(i), label = j)  

plt.title('Naive Bayes (Training set)')  
plt.xlabel('Feature 1')  
plt.ylabel('Feature 2')  
plt.legend()  
plt.show()
```
