```python
#train test split

from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
## standardizing the dataset
# for regularization we need to first do the standardization
from sklearn.preprocessing import StandardScaler
```

```python
scaler =  StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#cross validation
from sklearn.model_selection import cross_val_score 
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
#calculate mse
mse = cross_val_score (lr, X_train, y_train , scoring = 'neg_mean_squared_error', cv = 10)
```

```python
np.mean(mse)
```

```python
#prediction

y_pred = lr.predict(X_test)
```

```python
y_pred
```

```python
import seaborn as sns
sns.displot(y_pred - y_test, kind = 'kde') # in this here we got the left skewed 
```

```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(y_pred , y_test)
```

```python
score
```

# Ridge Regression


```python
from sklearn.linear_model import Ridge
```

```python
from sklearn.model_selection import GridSearchCV #GridSearchCV :  to search and find the optimal combination of hyperparameters for a given model.
```

```python
ridge_regressor = Ridge()
```

```python
ridge_regressor
```

```python
parameters = {'alpha' : [1,2,3,5,6,10,20, 30, 40, 50, 60, 70, 80,10]}
```

```python
ridgecv = GridSearchCV(ridge_regressor, parameters, scoring ='neg_mean_squared_error', cv = 5)
```

```python
ridgecv.fit(X_train, y_train)
```

```python
#to know the best parameter
print(ridgecv.best_params_)
```

```python
print(ridgecv.best_score_)
```

```python
ridge_pred = ridgecv.predict(X_test)
```

```python
import seaborn as sns

sns.displot(ridge_pred - y_test, kind = 'kde')
```

```python
score = r2_score(ridge_pred , y_test)
```

```python
score
```
