# Handling Missing Values for Categorical Data

```python
import pandas as pd
```

```python
df = pd.read_csv("titanic.csv")
```

```python
df
```

## 1. Frequent Category Imputation 

```python
most_frequent_category = df['Cabin'].mode()[0]
print(f"Most Frequent Category in 'Cabin': {most_frequent_category}")
```

```python
# Fill missing values in 'Cabin' with the most frequent category
df['Cabin'].fillna(most_frequent_category, inplace=True)
```

```python
df
```

## 2. Adding a new variable to captute Nan Values

```python
import numpy as np
```

```python
df1 = pd.read_csv("titanic.csv")
df1
```

```python
# Add a new column to indicate missing values in the 'Cabin' column
df1['Cabin_missing'] = df1['Cabin'].isnull().astype(int)
```

```python
df1
```

# 3.Replacing all NaN Values with a New Category

Mostly used this technique

here use fillna()

```python
df2 = pd.read_csv("titanic.csv")
```

```python
# Display the initial 'Cabin' column with missing values
print("Before replacing NaN values:")
print(df2['Cabin'].head(10))
```

```python
# Replace all NaN values in the 'Cabin' column with a new category 'Unknown'
df2['Cabin'] = df2['Cabin'].fillna('Unknown')
```

```python
# Display the DataFrame to show the updated 'Cabin' column
print("\nAfter replacing NaN values with 'Unknown':")
print(df2['Cabin'].head(10))
```
