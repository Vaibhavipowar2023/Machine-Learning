when we applying one hot encoding it creates 4209 new features. which may leads the curse of dimensionality which effect on model accuracy .

for overcome this problem we take most frequent "10"  most categories into features and apply one hot encoding and ignore rest of all the categories

```python
#lets find the top 10 most frequent categories for the variable  X2

df.X2.value_counts().sort_values(ascending = False).head(20)
```

```python
#lets make a list with the most frequent categories of the variable

top_10 = [X for X in df.X2.value_counts().sort_values(ascending = False).head(10).index]


#[X for X in ...]:This is a list comprehension that creates a list containing the top 10 most frequent categories.
#.index:This extracts the indices (which correspond to the original category names in X2) of these top 10 frequent values.
```

```python
top_10
```

```python
# and now we make the 1- binary variables

for label in top_10 : #This is a loop that iterates over each category in the top_10 list (which contains the top 10 most frequent categories from X2).
    df[label] = np.where(df['X2'] == label , 1, 0)

df[['X2']+top_10].head(40)


#for label in top_10 : this is loop that itrates over each category in the top_10 list

# df[label] = np.where(df['X2'] == label , 1, 0) :For each label (i.e., each category in the top 10), a new column is created in the DataFrame df.
# The np.where function is used to assign a value of 1 if the value in the X2 column matches the label, otherwise, it assigns a value of 0.
# Essentially, this creates a binary column for each of the top 10 categories, indicating the presence (1) or absence (0) of that category in each row

# df[['X2']+top_10].head(40)
# This line selects the X2 column along with the newly created binary columns corresponding to the top 10 categories.
# .head(40) returns the first 40 rows of this selection, allowing you to see the original X2 values alongside the new binary columns..
```

```python
#get whole set of dummy variables, for all the categorical variables

def one_hot_top_X(data, variable, top_x_labels):
    #functio to create the dummy variables for the most frequent labels
    #we can vary the number of most frequent lables that we encode
    
    for label in top_x_labels:
        df[variable + '_' + label] = np.where(data[variable]==label,1,0)
        
#read the data again
df = pd.read_csv("Mercedesbenz.csv", usecols =['X0','X1','X2','X3','X4','X5', 'X6' ,'X8'])

#encode X2 into the 10 most frequent categoris
one_hot_top_X(df, 'X2', top_10)

df.head()
```

```python
#find the 10 most frequent categories for X1

top_10 = [X for X in df.X1.value_counts().sort_values(ascending = False).head(10).index]
top_10
```
